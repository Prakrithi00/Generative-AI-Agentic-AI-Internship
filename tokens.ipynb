```python
text = "Machine Learning is amazing"
tokens = text.split()
longest = max(tokens, key=len)
print(longest)
```

```python
text = "I love learning NLP"
tokens = text.split()
for i, token in enumerate(tokens):
    print(i, token)
```

```python
text = "Machine Learning is amazing"
tokens = text.split()
shortest = min(tokens, key=len)
print(shortest)
```

```python
from collections import Counter
text = "AI AI AI NLP NLP"
tokens = text.split()
freq = Counter(tokens)
print(freq)
```

```python
text = "AI AI NLP NLP DeepLearning"
tokens = text.split()
unique_tokens = set(tokens)
print(unique_tokens)
```

```python
text = "I love AI and NLP"
tokens = text.split()
vocab = sorted(set(tokens))
print(vocab)
```

```python
text = "I love AI"
tokens = text.split()
vocab = {word:i for i,word in enumerate(set(tokens))}
token_ids = [vocab[word] for word in tokens]
print(token_ids)
```

```python
vocab = {
"I":0,
"love":1,
"AI":2
}
reverse_vocab = {v:k for k,v in vocab.items()}
ids = [0,1,2]
sentence = [reverse_vocab[i] for i in ids]
print(" ".join(sentence))
```

```python
class SimpleTokenizer:
    def tokenize(self, text):
        return text.split()

obj = SimpleTokenizer()
print(obj.tokenize("I love AI"))
```
